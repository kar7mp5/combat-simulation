```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

fig, ax = plt.subplots()

x = np.arange(51)
y = np.arange(51)

line, = ax.plot([], [], lw=2)

ax.set_xlim(0, 50)
ax.set_ylim(0, 50)

def update(frame):
    line.set_data(x[:frame], y[:frame])
    return line,

anim = animation.FuncAnimation(fig, update, frames=len(x) + 1,
                               interval=50, blit=True)

HTML(anim.to_jshtml())
```
